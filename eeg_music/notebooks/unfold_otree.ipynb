```python
import os
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
BASE_PATH = "../data/02_EYODF Pilot Data/Pilot (Same Songs)"
OTREE_PATH = os.path.join(BASE_PATH, "all_participants_otree.csv")

NUMBER_OF_TRACKS = 40
```

```python
meta = pd.read_csv(OTREE_PATH).iloc[0]
```

```python
meta = meta.reset_index()
meta.columns = ["otree_variable", "value"]
```

```python
for o in meta.otree_variable:
    print(o)
```

```python
meta = meta.assign(
    mode=meta.apply(lambda row: row.otree_variable.split(".")[0], axis=1),
)
meta = meta.assign(
    discovery_dimension=meta.loc[meta["mode"] == "Music_Discovery"].apply(lambda row: row.otree_variable.split(".")[-1], axis=1),
    track_id=meta.loc[meta["mode"] == "Music_Discovery"].apply(lambda row: int(row.otree_variable.split(".")[1]), axis=1)
)
```

```python
meta
```

```python
tracks = meta.loc[meta["mode"] == "Music_Discovery"][["value", "mode", "discovery_dimension", "track_id"]].pivot(
    index="track_id",
    columns="discovery_dimension",
    values="value"
)
tracks
```

```python
tracks = tracks.assign(
    time_start=pd.to_datetime(tracks.time_start, unit="ms"),
    time_add=pd.to_datetime(tracks.time_add.replace(0, np.nan), unit="ms", errors="coerce")
)
```
